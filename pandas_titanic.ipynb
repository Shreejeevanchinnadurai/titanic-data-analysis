```python
# Q1: How many rows and columns?
print(df.shape)
# Q2: Which columns have missing values?
print(df.isnull().sum())


# Q3: What is the average age of passengers?
print(df["Age"].mean())
# Q4: How many passengers survived vs died?

#     Hint: df['Survived'].value_counts()
print(df['Survived'].value_counts())

# Q5: How many male vs female passengers?
#     Hint: df['Sex'].value_counts()
print(df['Sex'].value_counts())
```

```python
# Step 1 — Fix Age (fill with median)
df['Age'] = df['Age'].fillna(df['Age'].median())

# Step 2 — Fix Embarked (fill with mode)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Step 3 — Drop Cabin (too many missing — 77%!)
df = df.drop(columns=['Cabin'])

# Step 4 — Verify no missing values left
print(df.isnull().sum())
```

```python
# Q1: Survival rate by Gender
print(df.groupby('Sex')['Survived'].mean())

# Q2: Survival rate by Passenger Class
print(df.groupby('Pclass')['Survived'].mean())

# Q3: Average age by Passenger Class
print(df.groupby('Pclass')['Age'].mean())

# Q4: Average fare by Passenger Class
print(df.groupby('Pclass')['Fare'].mean())
```

```python
# Q5: Survival rate by Embarked port

print(df.groupby('Embarked')['Survived'].mean())
print("====================================================")
# Q6: Average fare paid by male vs female
print(df.groupby('Sex')['Fare'].mean())
print("====================================================")
# Q7: How many passengers in each class?
print(df.groupby('Pclass')['Name'].count())
print("====================================================")
# Q8: Average age of survived vs died
print(df.groupby('Survived')['Age'].mean())
print("====================================================")
# Q9: Add AgeGroup column (Child/Adult/Senior)
df['AgeGroup']=df['Age'].apply(lambda x:'Child'if x<18 else ('Senior' if x>=40 else'Adult'))
print(df)
print("====================================================")
# Q10: Survival rate by AgeGroup
print(df.groupby('AgeGroup')["Survived"].mean())
```

Because most people were poor and couldn't
